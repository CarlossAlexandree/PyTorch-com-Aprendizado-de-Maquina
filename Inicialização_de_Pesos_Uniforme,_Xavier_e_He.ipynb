```python
import torch
import matplotlib.pyplot as plt

def plot_weights(weights, title):
    plt.hist(weights.numpy().flatten(), bins=50)
    plt.title(title)
    plt.show()

# Uniforme
w_uniform = torch.empty(1000).uniform_(-1, 1)
plot_weights(w_uniform, 'Uniforme (-1, 1)')

# Xavier
w_xavier = torch.empty(1000)
torch.nn.init.xavier_uniform_(w_xavier.view(1, -1))
plot_weights(w_xavier, 'Xavier Uniforme')

# He
w_he = torch.empty(1000)
torch.nn.init.kaiming_uniform_(w_he.view(1, -1), nonlinearity='relu')
plot_weights(w_he, 'He Uniforme')
```
