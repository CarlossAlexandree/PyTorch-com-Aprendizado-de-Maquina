```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

x = torch.linspace(-5, 5, 100)

# Funções de ativação
sigmoid = torch.sigmoid(x)
tanh = torch.tanh(x)
relu = F.relu(x)

# Derivadas
sigmoid_grad = sigmoid * (1 - sigmoid)
tanh_grad = 1 - tanh**2
relu_grad = torch.where(x > 0, torch.ones_like(x), torch.zeros_like(x))

# Plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, tanh, label='Tanh')
plt.plot(x, relu, label='ReLU')
plt.title('Funções de Ativação')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, sigmoid_grad, label='Sigmoid Derivada')
plt.plot(x, tanh_grad, label='Tanh Derivada')
plt.plot(x, relu_grad, label='ReLU Derivada')
plt.title('Derivadas')
plt.legend()
plt.show()
```
