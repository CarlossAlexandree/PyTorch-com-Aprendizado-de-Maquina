```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(momentum=0.0):
    x = 5
    v = 0
    trajectory = []

    for i in range(50):
        grad = 2 * x  # derivada de x^2
        v = momentum * v - 0.1 * grad
        x += v
        trajectory.append(x)

    return trajectory

plt.plot(gradient_descent(momentum=0.0), label='Sem Momentum')
plt.plot(gradient_descent(momentum=0.9), label='Com Momentum')
plt.title('Gradiente Descendente com e sem Momentum')
plt.xlabel('Iterações')
plt.ylabel('Posição')
plt.legend()
plt.show()
```
